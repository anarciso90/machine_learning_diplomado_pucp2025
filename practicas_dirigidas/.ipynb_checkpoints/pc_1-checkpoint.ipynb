## 1.2 Estrategias de Validación en Investigación Social

**Pregunta 1** 

*Análisis de la Calidad de la Estimación del Error*: Para que su artículo sea creíble, la socióloga quiere una estimación del error que sea lo más cercana posible al “verdadero” error de generalización (es decir, una estimación con bajo sesgo).

¿Qué método (A o B) produce una estimación menos sesgada y por qué?

**Respuesta** 

El método B (LOOCV) produce una estimación del error menos sesgada. Esto se debe a que se optimiza el número de casos de la base de datos y se realizan n números de procesos de validación dejando, en cada una de las iteraciones, una sola observación fuera para que sea usada para el error de prueba. En otras palabras, cada modelo se entrena con casi toda la muestra disponible, lo que reduce el sesgo en la estimación del error de generalización: su error estimado está muy cerca del error que tendría el modelo final entrenado con todos los datos. El método LOOCV, al generar n errores cuadráticos de predicción, se robustece el resultado de la estimación de error, acercándolo al "verdadero" error de generalización. 

**Pregunta 2**

*Recomendación Metodológica*: ¿Qué método le recomendaría a la socióloga para su publicación? Justifique su elección, argumentando por qué es la opción más equilibrada y defendible.

**Respuesta**

Pese a que el modelo B genera una estimación menos sesgada, recomendaría el método A (5-fold CV) como la opción más equilibrada y defendible. El método LOOCV hace un trade entre el sesgo y la varianza: si bien nos ofrece un bajo sesgo, nos da a cambio una alta varianza; lo cual significa que pequeñas diferencias en los datos pueden generar grandes variaciones en el error estimado, especialmente si el modelo es no lineal o de alta varianza (como árboles o redes neuronales). Además, el método de LOOCV es computacionalmente más pesado, ya que requiere, en este caso, que el modelo se entrene 600 veces.

Por otro lado, el método 5-fold CV genera resultados más equilibrados entre sesgo y varianza, a la par que es más estable computacionalmente. También se puede mencionar que este método permite reportar una media y un intervalo, lo que mejora la transparencia. Adicional a ello, es importante considerar que la investigadora es socióloga y, en contextos de investigación social, donde los datos pueden tener ruido y variabilidad, una estimación con menor varianza es más defendible en publicaciones académicas. Además, en publicaciones se valora también estimadores robustos y replicables (reproducibles para los revisores).

# 2. Descenso de Gradiente

Implementaremos el Descenso de Gradiente (**Batch Gradient Descente**) para encontrar el mínimo de una función de dos variables. La función a optimizar es:

                                                f(x,y) = x^2 + 2y^2 - 2xy + 2x - 8y + 10

El mínimo teórico se encuentra en (x = 2, y = 3). 

```python
#Importamos librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
```

## 2.1. Definicón de la Función y su Gradiente

```python
# a) Creamos una función en Python f(punto) que devuelva el valor de la función f(x, y).
def f(punto):
    """Calcular el valor de la función en un punto [x, y]."""
    x = punto[0]
    y = punto[1]

    # Función cuadrática: f(x,y) = x^2 + 2y^2 - 2xy + 2x - 8y + 10.

    return x**2 + 2*y**2 - 2*x*y + 2*x - 8*y + 10

# b) Creamos una función grad_f(punto) que devuelva el gradiente ∇f(x, y)
def grad_f(punto):
    """Calcula el gradiente [df/dx, df/dy] en un punto [x, y]."""

    x = punto[0]
    y = punto[1]

    # Derivada de df/dx = 2x
    df_dx = 2*x - 2*y + 2

    # Derivada de df/dy = 2y
    df_dy = 4*y - 2*x - 8
    
    return np.array([df_dx, df_dy])
```

## 2.2. Implementación del Algoritmo 

```python
# a) Creamos una función descenso_gradiente(punto_inicial, alpha, n_iteraciones) que implemente el bucle de optimización 
# y devuelva la trayectoria de los puntos.
def descenso_gradiente (punto_inicial, alpha, n_iteraciones):
    """Ejecuta el algoritmo de Descenso de Gradiente."""
    trayectoria = [punto_inicial]
    punto_actual = punto_inicial.copy()

    for i in range(n_iteraciones):
        # Llama a la función gradiente que define el Paso 1
        grad = grad_f(punto_actual)

        # Actualiza el punto
        punto_actual = punto_actual - alpha * grad

        # Guarda el nuevo punto
        trayectoria.append(punto_actual)

        return np.array(trayectoria)
```

## 2.3. Visualización y Verificación (caso base)

```python
# a) Definimos: punto inicial = np.array([-5.0, 10.0]), alpha = 0.1, n iteraciones = 50.

punto_inicial = np.array([-5.0, 10.0])
alpha = 0.1
n_iteraciones = 50
minimo_teorico = np.array([2.0, 3.0]) # <-- REEMPLAZAR con el mínimo asignado

# b) Ejecutamos la función y generamos un gráfico de contorno de f(x, y).

trayectoria = descenso_gradiente(punto_inicial, alpha, n_iteraciones)

# --- PREPARAMOS LA MALLA PARA EL GRAFICO DE CONTORNO ---
# <-- AJUSTA los limites de x e y para que tu trayectoria se vea bien
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X,  Y = np.meshgrid(x_vals, y_vals)
Z = f([X, Y]) # Llama a la funcion f que definiste en el Paso 1

# c) Sobre el gráfico, dibujamos la trayectoria del algoritmo, usamos flechas.
plt.figure(figsize=(12, 10))

# Dibuja las curvas de nivel
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis', alpha=0.6)
plt.clabel(contour, inline=True, fontsize=8)

# Dibuja la trayectoria con flechas
for i in range(len(trayectoria) - 1):
    x_start, y_start = trayectoria[i]
    dx = trayectoria[i+1, 0] - x_start
    dy = trayectoria[i+1, 0] - y_start
    plt.arrow(x_start, y_start, dx, dy,
              head_width=0.2, head_length=0.3,
              fc='red', ec='red', length_includes_head=True,
              label='Trayectoria' if i == 0 else "")
    
# Dibuja los puntos clave
plt.plot(punto_inicial[0], punto_inicial[1], 'go', markersize=10,
         label=f'Inicio: ({punto_inicial[0]}, {punto_inicial[1]})')
plt.plot(trayectoria[-1, 0], trayectoria[-1, 1], 'mo', markersize=10,
         label=f'Final: ({trayectoria[-1, 0]:.2f}, {trayectoria[-1, 1]:.2f})')
plt.plot(minimo_teorico[0], minimo_teorico[1], 'k*', markersize=15,
         label=f'Minimo teorico: ({minimo_teorico[0]}, {minimo_teorico[1]})')

#Configura el grafico
plt.title(f'Descenso de Gradiente (alpha = {alpha})', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.grid(True)
plt.legend()
plt.colorbar(contour, label='f(x,y)')

plt.show()
```

d) **Pregunta**: ¿El algoritmo converge visualmente hacia el mínimo teórico? **No.** El resultado del algoritmo arroja un resultado totalmente distinto. La flecha se desvía hacia otro lado.
